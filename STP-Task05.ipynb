4. Data Selection – Indexing and Slicing data

```python
df.iloc[:,1:6:2]
```

```python
df_sub=df[['Rank','Year','Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)']]
df_sub
```

5. Data Selection – Based on Conditional filtering

```python
df_rev300=df_sub[df_sub['Revenue (Millions)']>300]
df_rev300.head()
```

```python
df_long=df_sub[df_sub['Runtime (Minutes)']>100]
df_long.head()
```

```python
df_rated=df_sub[df_sub['Rating']>7]
df_rated.head()
```

```python
df_voted=df_sub[df_sub['Votes']>500000]
df_voted.head()
```

6. Groupby operations

```python
df_year=df_sub.groupby('Year')
df_year.count()
```

```python
df_r=df_sub[df_sub['Revenue (Millions)']>300]
df_r=df_r.groupby('Year')[['Rank']].count()
print(df_r)
```

7. Sorting operation


```python
df_sorted=df_sub[df_sub['Rating']>7].sort_values(by=['Rating','Votes','Revenue (Millions)'],ascending=[True,True,False])
df_sorted.head()
```

```python
df_sorted=df_sub[df_sub['Rating']>7].sort_values(by=['Rating','Votes','Revenue (Millions)'],ascending=[True,False,False])
df_sorted.head()
```
