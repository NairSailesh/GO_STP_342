4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
def sin(t):

    return np.sin(t)

t1 = np.arange(0.0, 16.0, 0.1)
plt.plot(t1, sin(t1), 'b-')
plt.show()
```

```python
def cos(t):

    return np.cos(t)

t1 = np.arange(0.0, 16.0, 0.1)
plt.plot(t1, sin(t1), 'b-',t1, cos(t1), 'r-')
plt.show()
```
